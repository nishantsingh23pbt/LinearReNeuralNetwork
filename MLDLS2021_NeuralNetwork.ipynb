```python
#importing lib
import tensorflow as tf
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('train2.csv')

#print shape of dataset with rows and columns
print(dataset.shape)
```

```python
X = dataset.iloc[:,-1].values
Y = dataset.iloc[:,19].values

df=pd.DataFrame(X,Y)
print(df)
```

```python
#defining a single layer; single neuron model
model = tf.keras.Sequential([keras.layers.Dense(units=3, activation='relu', input_shape=[1])])
model.add(Dense(1))
print(model.summary())
```

```python
tf.keras.utils.plot_model(
    model,
    to_file="model.png",
    show_shapes=True,
    show_dtype=False,
    show_layer_names=True,
    rankdir="TB",
    expand_nested=False,
    dpi=96,
)
```

```python

#defining the model evaluation for backpropogation
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
#taraining the model with iterations (No. of front & back propogation)
run_model = model.fit(X, Y, epochs=700)
```

```python
print(model.predict([40.0]))
```

```python
# list all data in history
print(run_model.history.keys())
```

```python
# summarize history for accuracy
plt.plot(run_model.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.show()
```
